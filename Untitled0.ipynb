```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/DS22_Avocado_Prices_Data.xlsx")
df.head()
```

```python
display(df.head())

num_rows, num_columns = df.shape
print("Shape (rows, columns):", df.shape)

duplicates = df.duplicated().sum()
print("Duplicate records:", duplicates)

missing_per_column = df.isnull().sum()
print("\nMissing values per column:")
print(missing_per_column)

```

```python
df.drop(columns=["Column 1"], errors='ignore',inplace=True)

df.head()

df['year'] = df['year'].astype('category')

df.head()

non_missing_data = df.dropna()
non_missing_data = non_missing_data.drop_duplicates()

df_cleaned = pd.concat([non_missing_data, df[df.isna().any(axis=1)]])

print("The shape of the dataset after removing the duplicate records is:",df_cleaned.shape)
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

if numeric_cols:
    correlation_matrix = df[numeric_cols].corr()

    plt.figure(figsize=(10, 6))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Feature Correlation Matrix")
    plt.show()

    print("\nCorrelation insights:")
    print("- High correlation (close to 1 or -1) means features are strongly related.")
    print("- Highly correlated independent variables can cause multicollinearity, reducing model accuracy in linear models (like regression).")
    print("- Tree-based models (Random Forest, XGBoost) are less sensitive to correlation, but redundant features may still slow training.")
else:
    print("No numeric columns available for correlation analysis.")

```

```python
print(df_cleaned.describe(include='all'))

categorical_cols = df_cleaned.select_dtypes(include=['object', 'category']).columns

for col in categorical_cols:
    print(f"{col}: {df_cleaned[col].nunique()} unique values")
```

```python
# Summary of numeric variables
print("\nSummary of numeric variables:")
print(df_cleaned.describe())

# Summary of categorical variables
categorical_cols = df_cleaned.select_dtypes(include=['object','category']).columns
print("\nSummary of categorical variables:")
for col in categorical_cols:
    print(f"{col}: {df_cleaned[col].nunique()} unique values")
    if df_cleaned[col].nunique() <= 10:
        print("  Values:", df_cleaned[col].unique())
```

```python
import numpy as np

numeric_cols = df_cleaned.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
    col_range = (df_cleaned[col].min(), df_cleaned[col].max())
    col_mean = df_cleaned[col].mean()
    col_median = df_cleaned[col].median()
    print(f"\n{col}:")
    print(f"  Range: {col_range}")
    print(f"  Mean: {col_mean:.2f}")
    print(f"  Median: {col_median:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of AveragePrice
plt.figure(figsize=(8,5))
sns.histplot(df_cleaned['AveragePrice'], bins=30, kde=True)
plt.title("Distribution of Avocado Average Prices")
plt.xlabel("Average Price ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Boxplot by type
plt.figure(figsize=(8,5))
sns.boxplot(x="type", y="AveragePrice", data=df_cleaned)
plt.title("Price Comparison: Conventional vs Organic Avocados")
plt.xlabel("Avocado Type")
plt.ylabel("Average Price ($)")
plt.show()
```

```python
# Top 10 regions by total volume
plt.figure(figsize=(12,6))
region_sales = df_cleaned.groupby("region")["Total Volume"].sum().sort_values(ascending=False).head(10)
region_sales.plot(kind="bar")
plt.title("Top 10 Regions by Total Avocado Volume")
plt.xlabel("Region")
plt.ylabel("Total Volume")
plt.show()
```

```python
earliest_date = df_cleaned['Date'].min()
print("\nEarliest Date recorded:", earliest_date)

before_2010 = df_cleaned[df_cleaned['Date'] < "2010-01-01"]
print("Any data before 2010?:", not before_2010.empty)
```

```python
idx_max_price = df_cleaned['AveragePrice'].idxmax()
highest_price = df_cleaned.loc[idx_max_price, 'AveragePrice']
highest_price_region = df_cleaned.loc[idx_max_price, 'region']
print("\nHighest Average Price:", highest_price, "from region:", highest_price_region)
```

```python
idx_max_volume = df_cleaned['Total Volume'].idxmax()
highest_volume = df_cleaned.loc[idx_max_volume, 'Total Volume']
highest_volume_region = df_cleaned.loc[idx_max_volume, 'region']
print("Highest Total Volume:", highest_volume, "from region:", highest_volume_region)
```
